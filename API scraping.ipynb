```python
necessary_columns = ['compnumber','reportid','mapcode','reporttype','reportdate','longname',
                     'amount','status','country','region','indicator','statement']
```

The above given columns are necessary for the following reasons:
1. compnumber : company number, which makes it easy to recognise the company
2. report id : the id number of the report of the financial report 
3. mapcode : code represents a financial metric 
4. amount : the amount of the financial metric 
5. report date : the date the report is made 
6. report type: the type of report
7. status: status of the financial report 
8. country: the country of the company 
9. region: the region the country is located in 
10. indicator : financial indicator reported 
11. statement: statement derived from the statement


```python
df = df_metric[necessary_columns]
```

```python
filtered_df  = df[df['indicator'] == 'Accrued Expenses Turnover']
```

```python
filtered_df.describe()
```

```python
filtered_df['country'].value_counts()
```

```python
#cleaning country names
def names(name):
    if name == 'USA':
        return 'United States of America'
    elif name == 'DEU':
        return 'Germany'
    elif name == 'JPN':
        return 'Japan'
    elif name == 'CYM':
        return 'Cayman Islands'
    elif name == 'BHS':
        return 'Bahamas'
    elif name == 'IRL':
        return 'Ireland'
```

```python
filtered_df['country'] = filtered_df['country'].apply(names)
```

```python
filtered_df = filtered_df.rename(columns = {'longname':'company_name'})
```

```python
filtered_df['country'].value_counts()
```

```python
#understanding financial trends over time
# # Filter for the time period 2010 to 2015
filtered_df['reportdate'] = pd.to_datetime(filtered_df['reportdate'])
filtered_df = filtered_df[(filtered_df['reportdate'].dt.year >= 2010) & (filtered_df['reportdate'].dt.year <= 2015)]

# Selecting relevant columns. Assuming 'company_name' is the column with company names.
# Replace 'company_name', 'reportdate', and 'metric_value_column' with actual column names.
relevant_data = filtered_df[['company_name', 'reportdate', 'amount']].copy()
```

```python
# Plotting the trend
plt.figure(figsize=(12, 6))

for company in relevant_data['company_name'].unique():
    company_data = relevant_data[relevant_data['company_name'] == company]
    plt.plot(company_data['reportdate'], company_data['amount'], label=company)

plt.title('Trend Analysis of Accrued Expenses Turnover (2010-2015)')
plt.xlabel('Report Date')
plt.ylabel('Accrued Expenses Turnover')
plt.xticks(rotation=45)
plt.legend(title='Company', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#grouping data by country and calculating the average of the amount 
country_avg = filtered_df.groupby('country')['amount'].mean()
```

```python
plt.figure(figsize=(12,6))
country_avg.sort_values(ascending=False).plot(kind='bar')
plt.title('Average Financial Metric by Country')
plt.xlabel('Country')
plt.ylabel('Average Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.show()
```
